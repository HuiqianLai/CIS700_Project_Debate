```python
import pandas as pd

# 加载CSV文件
csv_path = 'motion2_csv.csv'
df = pd.read_csv(csv_path)

# 打开txt文件准备写入
output_path = 'motions_output.txt'
with open(output_path, 'w', encoding='utf-8') as f:
    # 遍历每一行，写入ID和Motion
    for index, row in df.iterrows():
        motion_id = row['ID']
        motion_text = row['Motion']
        f.write(f"{motion_id}. {motion_text}\n\n")

print(f"数据已成功保存到 {output_path}")

# 同时在控制台打印前几条查看效果
print("\n前5条数据预览：")
for index, row in df.head().iterrows():
    print(f"{row['ID']}. {row['Motion']}")
```
